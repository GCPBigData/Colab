# Janela deslizante (sliding) x Janela expandida (expanding)

[2003, 2004, 2005, 2006, 2007] -> [2003, 2004, 2005, 2006] [2007]  
[2003, 2004, 2005, 2006, 2007] -> [2006, 2007] [2008]

```python
periodos = sorted(rjm['mdct'].dt.year.value_counts().index[1:])
```

```python
rjm_sorted = rjm.sort_values("mdct")
```

```python
rjm_sorted.head()
```

```python
rjm_sorted['y'] = rjm_sorted['temp'].shift(-1)
```

```python
rjm_sorted = rjm_sorted.iloc[:-1].fillna(0.)
```

```python
colunas = rjm_sorted.columns[14:-1]
```

```python
colunas
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
```

```python
erro_retreino = []
for ano_treino in periodos[1:-1]:
    Xtreino = rjm_sorted[rjm_sorted['mdct'].dt.year == ano_treino][colunas]
    Xval = rjm_sorted[rjm_sorted['mdct'].dt.year == ano_treino+1][colunas]
    ytreino = rjm_sorted[rjm_sorted['mdct'].dt.year == ano_treino]['y']
    yval = rjm_sorted[rjm_sorted['mdct'].dt.year == ano_treino+1]['y']
    
    mdl = RandomForestRegressor(n_estimators=100, n_jobs=-1, random_state=0)
    mdl.fit(Xtreino, ytreino)
    
    p = mdl.predict(Xval)
    erro = np.sqrt(mean_squared_error(yval, p))
    print("Ano treino", ano_treino, 'Ano valid', ano_treino + 1, "Erro", erro)
    erro_retreino.append(erro)
```
