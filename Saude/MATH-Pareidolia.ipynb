##Elliott
O princípio de Elliott Wave postula que a psicologia do investidor coletivo, ou psicologia da multidão , se move entre o otimismo e o pessimismo em sequências naturais. Essas oscilações de humor criam padrões evidenciados nos movimentos de preços dos mercados em todos os graus de tendência ou escala de tempo.

No modelo de Elliott, os preços de mercado alternam entre uma fase impulsiva ou motora e uma fase corretiva em todas as escalas de tempo de tendência, como mostra a ilustração. Os impulsos são sempre subdivididos em um conjunto de 5 ondas de grau inferior, alternando novamente entre o motivo e o caráter corretivo, de modo que as ondas 1, 3 e 5 são impulsos e as ondas 2 e 4 são retraços menores das ondas 1 e 3. Ondas corretivas subdividir em 3 ondas de menor grau começando com um impulso de contra-tendência de cinco ondas, um retrace e outro impulso. Em um mercado baixista, a tendência dominante é de baixa, então o padrão é inverso - cinco ondas para baixo e três para cima. As ondas motrizes sempre se movem com a tendência, enquanto as ondas corretivas se movem contra ela.

#Padrão de cinco ondas (tendência dominante)

Onda 1: a Onda um raramente é óbvia em seu início. Quando a primeira onda de um novo mercado altista começa, as notícias fundamentais são quase universalmente negativas. A tendência anterior é considerada ainda fortemente em vigor. Os analistas fundamentais continuam revisando suas estimativas de lucros para baixo; a economia provavelmente não parece forte. As pesquisas de opinião são decididamente de baixa, as opções de venda estão em voga e a volatilidade implícita no mercado de opções é alta. O volume pode aumentar um pouco com o aumento dos preços, mas não o suficiente para alertar muitos analistas técnicos.

Onda 2: a onda dois corrige a onda um, mas nunca pode se estender além do ponto inicial da onda um. Normalmente, as notícias ainda são ruins. À medida que os preços testam novamente a mínima anterior, o sentimento de baixa cresce rapidamente, e "a multidão" arrogantemente lembra a todos que o mercado de baixa ainda está profundamente arraigado. Ainda assim, alguns sinais positivos aparecem para quem está procurando: o volume deve ser menor durante a onda dois do que durante a onda um, os preços geralmente não remontam mais de 61,8% (ver seção Fibonacci abaixo) dos ganhos da onda um e os preços devem cair um padrão de três ondas.

Onda 3: A onda três é geralmente a maior e mais poderosa em uma tendência (embora algumas pesquisas sugiram que, nos mercados de commodities, a onda cinco é a maior). A notícia agora é positiva e analistas fundamentais começam a elevar as estimativas de lucro. Os preços sobem rapidamente, as correções têm vida curta e são superficiais. Qualquer um que esteja procurando "entrar em um recuo" provavelmente perderá o barco. Quando a onda três começa, as notícias provavelmente ainda são de baixa e a maioria dos participantes do mercado permanece negativa; mas no ponto médio da onda três, "a multidão" frequentemente se junta à nova tendência de alta. A onda três frequentemente estende a onda um em uma proporção de 1,618: 1 .

Onda 4: A onda quatro é claramente corretiva. Os preços podem desviar-se para os lados por um longo período, e a onda quatro normalmente remonta a menos de 38,2% da onda três (veja as relações de Fibonacci abaixo). O volume está bem abaixo do da onda três. Este é um bom lugar para comprar um retrocesso se você entende o potencial à frente para a onda 5. Ainda assim, as quartas ondas costumam ser frustrantes devido à falta de progresso na tendência maior.

Onda 5: a onda cinco é a perna final na direção da tendência dominante. As notícias são quase universalmente positivas e todos estão otimistas. Infelizmente, é quando muitos investidores médios finalmente aceitam, antes do topo. O volume é freqüentemente menor na onda cinco do que na onda três, e muitos indicadores de momentum começam a mostrar divergências (os preços atingem uma nova máxima, mas os indicadores não atingem um novo pico). No final de um grande mercado altista, os ursos podem muito bem ser ridicularizados (lembre-se de como as previsões de um topo no mercado de ações em 2000 foram recebidas).



##Padrão de três ondas (tendência corretiva)

Onda A: Normalmente, as correções são mais difíceis de identificar do que os movimentos de impulso. Na onda A de um mercado em baixa, as notícias fundamentais geralmente ainda são positivas. A maioria dos analistas vê a queda como uma correção em um mercado altista ainda ativo. Alguns indicadores técnicos que acompanham a onda A incluem aumento de volume, aumento da volatilidade implícita nos mercados de opções e, possivelmente, uma curva mais alta nos contratos em aberto nos mercados de futuros relacionados .

Onda B: os preços se invertem, o que muitos veem como uma retomada do mercado de alta, agora há muito extinto. Aqueles familiarizados com a análise técnica clássica podem ver o pico como o ombro direito de um padrão de reversão de cabeça e ombros. O volume durante a onda B deve ser menor do que na onda A. Neste ponto, os fundamentos provavelmente não estão mais melhorando, mas provavelmente ainda não ficaram negativos.

Onda C : os preços caem impulsivamente em cinco ondas. O volume aumenta e, na terceira etapa da onda C, quase todo mundo percebe que um mercado em baixa está firmemente estabelecido. A onda C é tipicamente pelo menos tão grande quanto a onda A e frequentemente se estende a 1,618 vezes a onda A ou além.

##Uma contagem correta de ondas de Elliott deve observar três regras:

A onda 2 nunca refaz mais do que 100% da onda 1.

A onda 3 não pode ser a mais curta das três ondas de impulso, a saber, ondas 1, 3 e 5.

A onda 4 não se sobrepõe ao território de preço da onda 1, exceto no raro caso de formação de um triângulo diagonal.

```
# sqrt para calcular raiz quadrada
from math import sqrt 
n = 3
def fibonacci(n):
    if n < 2:
        return n
    else:
        u = ((1+sqrt(5))/2)
        j = ((u**n-(1-u)**n)/sqrt(5))
        # round nos permite aredondar numeros flutuantes
        #return round(j)
        return j 
```

```
for x in range(369):
    print(fibonacci(x))
```
